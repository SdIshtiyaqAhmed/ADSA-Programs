Problem Statement:

Simulate Recursive Function Calls Using a Stack. Use Factorial or Fibonacci as an Example. Display The Function Call Stack.

Solution:

Factorial Example

```python
# Factorial using recursion

def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n-1)

num = int(input("Enter a number: "))
print(factorial(num))
```

```python
# Factorial using iteration

call_stack = []
num = int(input("Enter a number: "))
print("Call Stack: ",call_stack)
call_stack.append(["CALL", {"n":num-1}])

while(call_stack):
  print("Call Stack: ",call_stack)
  action, data = call_stack.pop()
  n = data["n"]

  if action == "CALL":
    if n == 1:
      call_stack.append(["MULTIPLY",{"n":n}])
    else:
      call_stack.append(["CALL", {"n":n-1}])
      call_stack.append(["MULTIPLY",{"n":n}])

  elif action == "MULTIPLY":
    num *= n
print(num)
```

Fibonacci Example

```python
# fibonacci using recursion

def fibonacci(n):
  if n <= 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)

num = int(input("Enter a number: "))
print(fibonacci(num))
```

```python
# fibonacci using iteration

call_stack = []
num = int(input("Enter a number: "))
results = {}

if num == 1:
  print(num)

else:
  for i in range(1,num+1):
    print("Call Stack: ",call_stack)
    call_stack.append(["CALL", {"n":i}])

  while(call_stack):
    print("Call Stack: ",call_stack)
    action, data = call_stack.pop()
    n = data["n"]

    if n in results:
      continue

    if action == "CALL":
      if n <= 2:
        results[n] = n
      else:
        call_stack.append(["ADD",{"n":n}])

        call_stack.append(["CALL",{"n":n-1}])
        call_stack.append(["CALL",{"n":n-2}])

    elif action == "ADD":
      results[n] = results[n-1] + results[n-2]

print(results.get(num))
```
