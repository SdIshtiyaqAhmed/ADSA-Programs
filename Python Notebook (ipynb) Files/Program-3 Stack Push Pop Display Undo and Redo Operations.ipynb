Problem Statement:

Create an Undo-Redo System Using Two Stacks. Implement Push, Pop, and Display Operations.

Solution:

```python
# define the stack structure and operations
class Stack:

    # initialize the stack
    def __init__(self):
        self.stack = []

    # push element into the stack
    def push(self, element):
        self.stack.append(element)

    # pops and returns the topmost element from the stack
    def pop(self):
        if len(self.stack) == 0:
            return

        else:
            element = self.stack[-1]
            self.stack.pop()

        return element

    # returns the topmost element from the stack
    def peek(self):
        if len(self.stack) == 0:
            return None
        else:
            return self.stack[-1]

    # displays all the elements in the stack
    def display(self):
        for i in self.stack:
            print(i, end=" ")
        print()

user_prompt = """
Choose any of the following:
1. Push, 2. Pop, 3. Display, 4. Undo, 5. Redo, 6. Exit
"""

stack = Stack()
undo = Stack()
redo = Stack()

while True:
    print(user_prompt)
    option = int(input("Enter your option: "))

    if option == 1: # push an element
        element = input("Enter an element: ")
        stack.push(element)

        # push the operation "push" and element into the undo stack
        undo.push(["push", element])

        # re-initialize the redo stack
        redo.stack = []

    elif option == 2: #pop
        element = stack.pop()

        # if the stack is not empty, then push the operation "pop" and the element into the undo stack
        if element != None:
            undo.push(["pop", element])

        # re-initialize the redo stack
        redo.stack = []

    elif option == 3:

        # display all the elements in the stack
        stack.display()

    elif option == 4:

        # if the undo stack is not empty
        if undo.peek() != None:

            # if the topmost element in the undo stack has the "push" operation
            # then pop an element from the stack
            if undo.peek()[0] == "push":
                stack.pop()

            # if the topmost element in the undo stack has the "pop" operation
            # then push an element into the stack
            elif undo.peek()[0] == "pop":
                stack.push(undo.peek()[1])

            # push the popped [operation, element] from the undo stack into the redo stack
            redo.push(undo.pop())

    elif option == 5:

        # if the redo stack is not empty
        if redo.peek() != None:

            # if the topmost element in the redo stack has the operation "push"
            # then push the topmost element from the redo stack into the stack
            if redo.peek()[0] == "push":
                stack.push(redo.peek()[1])

            # if the topmost element in the redo stack has the operation "pop"
            # then pop the topmost element from the stack
            elif redo.peek()[0] == "pop":
                stack.pop()

            # pop [operation, element] from the redo stack and push it into the undo stack
            undo.push(redo.pop())

    else:
        break
```
